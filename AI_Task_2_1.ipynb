```python
#Fibonacci Series of a given number
n = int(input("How many numbers? "))

x1, x2 = 0, 1
c = 0
if n <= 0:
   print("-ve numbers are not valid inputs") 
elif n == 1:
   print("Fibonacci sequence upto",n,":")
   print(x1)
else:
   print("Fibonacci sequence is :")
   while c < n:
       print(x1)
       x3 = x1 + x2
       x1 = x2
       x2 = x3
       c+= 1
```
